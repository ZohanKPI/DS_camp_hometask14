```python
from sklearn.datasets import make_blobs
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering, DBSCAN, KMeans

X, y = make_blobs(n_samples = 500, n_features = 2, centers = 5, 
                           cluster_std = 0.6, random_state = 0)
colors = np.array([plt.cm.Spectral(val)
          for val in np.linspace(0, 1, len(set(y)))])
plt.figure(figsize=(8,6))

plt.scatter(X[:,0], X[:,1], c= colors[y], s= 20)  
```

```python
agg_clustering_complete = AgglomerativeClustering(n_clusters=5, linkage='complete')
labels_complete = agg_clustering_complete.fit_predict(X)

agg_clustering_single = AgglomerativeClustering(n_clusters=5, linkage='single')
labels_single = agg_clustering_single.fit_predict(X)

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.scatter(X[:, 0], X[:, 1], c=colors[labels_complete], s=20)
plt.title("Agglomerative Clustering (Complete Linkage)")

plt.subplot(1, 2, 2)
plt.scatter(X[:, 0], X[:, 1], c=colors[labels_single], s=20)
plt.title("Agglomerative Clustering (Single Linkage)")

plt.show()
```

```python
dbscan = DBSCAN(eps=0.5, min_samples=5)
labels_dbscan = dbscan.fit_predict(X)

plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=colors[labels_dbscan], s=20)
plt.title("DBSCAN Clustering")
plt.show()
```

```python
import warnings
warnings.filterwarnings('ignore')

sse = []
k_range = range(1, 11)

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=0)
    kmeans.fit(X)
    sse.append(kmeans.inertia_)

plt.figure(figsize=(8, 6))
plt.plot(k_range, sse, marker='o')
plt.title("Elbow Method for K-Means Clustering")
plt.xlabel("Number of clusters (k)")
plt.ylabel("Sum of squared distances")
plt.xticks(k_range)
plt.show()
```
